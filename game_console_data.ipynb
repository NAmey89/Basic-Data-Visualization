```python
import pandas as pd
reviews = pd.read_csv('F:\\DataScience\\Datasets\\ign.csv')
reviews.head()
```

```python
reviews.shape
```

```python
### Indexing dataframes in pandas using iloc
### iloc allows us to retrieve rows and columns by position 

reviews.iloc[0:5,:]

### Above code is equivalent to reviews.head()
### 0:5 are index values of rows and it stops at 5th value but does not include it
```

```python
### Now that we know how iloc works we can remove the first column which doesn't have much useful info.

reviews = reviews.iloc[:,1:]
reviews.head()
```

```python
### For retrieving rows and columns by label use loc instead of iloc (indexing using labels)

reviews.loc[0:5,:]
```

```python
### Retrieving column by label

reviews.loc[0:5,'score']

### reviews.loc[0:5,['score']] <- use this if you want the column name to be shown
```

```python
### Retrieving multiple columns

### Note: since loc retrieves by label it returns the 5th row also 

reviews.loc[0:5,['score','score_phrase']]
```

```python
### scores above 7 and platform PC

filtered_scores = reviews[(reviews['score'] > 7) & (reviews['platform'] == 'PC')]
filtered_scores.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
### Total count for each platform

pf_ct = reviews['platform'].value_counts()
pf_ct
```

```python
### Scores greater than 7 for each platform

gt7_ct = reviews[reviews['score'] > 7]['platform'].value_counts()
gt7_ct
```

```python
### Histogram for scores of PC platform games

reviews[reviews['platform'] == 'PC']['score'].plot(kind = 'hist')
```

```python
### Graph of values of score_phrase for PlayStation 4

reviews['score_phrase'][reviews['platform'] == 'PlayStation 4'].value_counts().plot(kind = 'bar')
```

```python
### For Xbox 360

reviews['score_phrase'][reviews['platform'] == 'Xbox 360'].value_counts().plot(kind = 'bar')
```

```python
### Avg score of games for each platform

reviews.groupby(['platform'])['score'].mean()
```

```python
### Android platform scores

reviews[reviews['platform'] == 'Android']['score'].plot(kind = 'hist')
```

```python
### Actual count of values for PC platform

reviews['score_phrase'][reviews['platform'] == 'PC'].value_counts()
```
