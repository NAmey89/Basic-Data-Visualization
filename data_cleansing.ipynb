```python
### change data type of a pandas series

import pandas as pd
df_drinks = pd.read_csv('http://bit.ly/drinksbycountry')
df_drinks
```

```python
df_drinks.shape
```

```python
df_drinks.dtypes
```

```python
df_drinks['beer_servings'] = df_drinks.beer_servings.astype(float)

### converting a column type is useful when an int/float column is stored as a string/object
### by converting we can do mathematical calculations on the column
```

```python
df_drinks.dtypes
```

```python
df_drinks = pd.read_csv('http://bit.ly/drinksbycountry', dtype = {'wine_servings':float})
df_drinks.dtypes

### useful when we want to change the dtype at the time of importing the dataset
```

```python
df_drinks.groupby(['continent']).max()
```

```python
df_drinks.groupby(['continent']).min()
```

```python
df_drinks[df_drinks['continent'] == 'Asia']
```

```python
df_drinks.groupby(['continent'])['beer_servings', 'wine_servings'].mean().plot(kind = 'bar')
```
